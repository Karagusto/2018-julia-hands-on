### Split Apply Combine

    Splitting a dataset into groups, -- Groupby()
    Applying functions to each group -- by()

    Combining the results


    groupby()
    syntax=> groupby(DataFrame,[:column])



```julia
groupby(df,:species)
```


#### Split and Apply Function

    by()
    syntax=> by(DataFrame,:column,fxn)
    aggregate()
    syntax=> aggregate(DataFrame,:column,[fxns1,fxn2])



```julia
by(df,:species, size)
```

#### Statistics

```julia
by(df,:species,df-> mean(df[:sepal_length]))
```

```julia
by(df,:species) do x
    DataFrame(Mean = mean(x[:sepal_length]),
    STD = std(x[:sepal_length]))
end
```

```julia
aggregate(df,:species,[sum,mean,std,var])
```

# Data Visualization

```julia
Pkg.add("Gadfly")
using Gadfly
```

```julia
# to plot sepal_length against sepal_width
plot(df, x=:sepal_length, y=:sepal_width, Geom.point)
```

```julia
# changing the types of plot
plot(df, x=:sepal_length, y=:sepal_width,
         Geom.point, Geom.line)
```

```julia
# changing the color
plot(df, x=:sepal_length, y=:sepal_width,  color=:species,
         Geom.point)
```

## Working with different scales

```julia
Pkg.add("RDatasets")
using RDatasets
```

```julia
# plotting datasets without defining the scale
mammals = dataset("MASS", "mammals")
plot(mammals, x=:Body, y=:Brain, label=:Mammal, Geom.point, Geom.label)
```

```julia
# putting both axes in log-scale things look better
plot(mammals, x=:Body, y=:Brain, label=:Mammal,
         Geom.point, Geom.label, Scale.x_log10, Scale.y_log10)
```

```julia
#### Putting plots together
fig1a = plot(df, x="sepal_length", y="sepal_width", Geom.point)
fig1b = plot(df, x="sepal_width", Geom.bar)
fig1 = hstack(fig1a, fig1b)
```

```julia
draw(SVG("output.svg", 6inch, 3inch), plot([sin, cos], 0, 25))
```

```julia
draw(SVG("figure1.svg", 6inch, 3inch), fig1)
```
